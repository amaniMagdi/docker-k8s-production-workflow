> **Note**:
> 
> -   `RUN`: The command is triggered while building the Docker image.
> -   `CMD`: The command is triggered when we launch the created Docker image.

**Docker Run with Port Mapping**

`docker run -p 5000:8080 <image name/id>`

**1st port:** this will route the incoming requests to this port (5000) on localhost... 

**2nd port:** ...to this port (8080) inside the container.
